### Exercise 3

The condition that characterize the optimal amount of cake to leave for the next period in each period $\{W_2(period 1),W_3(period 2),W_4(period 3)\}$ are: 
$\begin{align*}
&\max_{W_2\in[0,W_1]}(u(W_1-W_2)+\max_{W_3\in[0,W_2]}\beta (u(W_2-W_3)+\max_{W_4\in[0,W_3]}\beta u(W_3-W_4)),(1)\\
&\max_{W_3\in[0,W_2]}(u(W_2-W_3)+\max_{W_4\in[0,W_3]}\beta u(W_3-W_4)),(2)\\
&and \max_{W_4\in[0,W_3]} u(W_3-W_4)(3).
\end{align*}$

Then we have $W_4 = 0$. To calculate $W_2$ and $W_3$, we take the derivatives of (1) and (2) and get $\begin{align*}
0 &= u'(1-W_2)+\beta(u'(W_2-W_3)+\beta u'W_3)\\
0 &= u'(W_2-W_3)+\beta u'W_3
\end{align*}$ 

Therefore, since we know $\beta= 0.9$ and the period utility function is $\ln(c_t)$, we have $W_2 = 0.631$ and $W_3 = 0.299$. Then we get $c_1 = 0.299$, $c_2 = 0.332$, and $c_3 = 0.299$. How $\{c_t\}^3_{t=1}$ and
$\{W_t\}^4_{t=1} evolve over the three periods could be shown as:

```python
import matplotlib.pyplot as plt

beta = 0.9
W = [1, 1-1/(1+beta+beta**2),1-(1+beta)/(1+beta+beta**2),0]
c = [(W[i]-W[i+1]) for i in range(len(W)-1)]
t = [0,1,2,3]

plt.plot(t,W)
plt.title("W")
plt.xlabel("t")
plt.ylabel("W")
plt.show()

plt.plot(t[1:],c)
plt.title("c")
plt.xlabel("t")
plt.ylabel("c")
plt.show()
```

### Exercise 5

We have $V_{T-1}(\bar{W})=u(\bar{W}-\psi_{T-1}\bar{W}))+\beta u(\psi_{T-1}(\bar{W}))$. Then taking the derivative, we have $-u'(\bar{W}-\psi_{T-1}(\bar{W}))+\beta u'(\psi_{T-1}(\bar{W}))=0$. Then we get $\psi_{T-1}(\bar{W}) = \frac{\beta}{1+\beta}\bar{W}$.

We get $V_{T-1}(\bar{W})=\ln((1-\frac{\beta}{1+\beta})\bar{W})+\beta\ln(\frac{\beta}{1+\beta}\bar{W})=\ln(\frac{1}{1+\beta}\bar{W})+\beta\ln(\frac{\beta}{1+\beta}\bar{W})$. Since $V_T(\bar{W})=u(\bar{W})=\ln(\bar{W})$ and $\psi_{T}(\bar{W})=0$, we have $V_T(\bar{W})\neq V_{T-1}(\bar{W})$ and $\psi_{T}(\bar{W})\neq \psi_{T-1}(\bar{W})$.

### Exercise 6

We have the finite horizon Bellman equation for the
value function at time $T-2$ is $\begin{align*} 
V_{T-2}(W_{T-2})\equiv&\max_{W_{T-1}}\ln(W_{T-2}-W_{T-1})+\beta V_{T-1}(W_{T-1})\\
\equiv&\max_{W_{T-1}}\ln(W_{T-2}-W_{T-1})+\beta(\ln(\frac{1}{1+\beta}W_{T-1})+\beta\ln(\frac{\beta}{1+\beta}W_{T-1}))\\
\equiv&\max_{W_{T-1}}\ln(W_{T-2}-W_{T-1})+\beta\ln(\frac{1}{1+\beta}W_{T-1})+\beta^2\ln(\frac{\beta}{1+\beta}W_{T-1})\\
\equiv&\max_{W_{T-1}}\ln(W_{T-2}-W_{T-1})+(\beta+\beta^2)\ln(W_{T-1})+\beta\ln(\frac{1}{1+\beta})+\beta^2\ln(\frac{\beta}{1+\beta})
\end{align*}$ 

Since we have $W_{T-1}=\psi_{T-2}(W_{T-2})$, we take the derivative and get $$-\frac{1}{W_{T-2}-\psi_{T-2}(W_{T-2})}+\frac{\beta+\beta^2}{\psi_{T-2}(W_{T-2})}=0.$$

Therefore, the analytical solutions are $\begin{align*}
\psi_{T-2}(W_{T-2})&=\frac{\beta+\beta^2}{1+\beta+\beta^2}W_{T-2}\\
V_{T-2}(W_{T-2})&=\ln(\frac{W_{T-2}}{1+\beta+\beta^2})+(\beta+\beta^2)\ln(\frac{(\beta+\beta^2)W_{T-2}}{1+\beta+\beta^2})+\beta\ln(\frac{1}{1+\beta})+\beta^2\ln(\frac{\beta}{1+\beta})\\
&=\ln(\frac{W_{T-2}}{1+\beta+\beta^2})+\beta\ln(\frac{\beta W_{T-2}}{1+\beta+\beta^2})+\beta^2\ln(\frac{\beta^2 W_{T-2}}{1+\beta+\beta^2})
\end{align*}$

### Exercise 10

```python
def u(c):
    new_c = np.log(c)
    return new_c

beta = 0.9
```

```python
W0 = np.zeros(100)
V0 = np.zeros(100)
u_p = u(W-W0)
V_p = u_p+beta*W0
```

### Exercise 14

```python
c_mat = W.reshape(-1,1)-W
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-10
u_mat = u(c_mat)

min_dist = 1e-10
Iter = 0
stop_iter = 1000
V0 = u_p

while Iter < stop_iter and dist > min_dist:
    V = np.tile(V0.reshape((1,100)),(100,1))
    V[~c_pos] = -9e+4
    new_V = (u_mat+beta*V).max(axis = 1)
    dist = np.sum((new_V-V0)**2)
    V0 = new_V
    W_index = np.argmax(u_mat+beta*V,axis = 1)
    Iter += 1
    
print(Iter, dist)
print("After "+str(Iter)+" iterations, the function converges to the fixed point.")
```

### Exercise 16

```python
from scipy.stats import norm

sigma = 0.5
mu = 4*sigma
M = 7
epsilon = np.linspace(mu-3*sigma,mu+3*sigma,M)
PDF = lambda x: norm(loc = mu, scale = sigma).pdf(x)
pdf = PDF(epsilon)
plt.plot(epsilon, pdf)
plt.show()
```

### Exercise 17

```python
c_mat = W.reshape(-1,1)-W
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-10
u_mat = u(c_mat)

d3dim = np.array([u_mat*e for e in epsilon])

V0 = np.zeros((100, M))
EV = V0 @ pdf.reshape((M,1)) 
EV_mat = np.tile(EV.reshape((1,100)),(100,1))
EV_mat[~c_pos] = -9e+4
EV_3d = np.array([EV_mat for i in range(M)])

V_3d = d3dim + beta*EV_3d
V_new = np.zeros((100,M))
W_prime = np.zeros((100,M))

for i in range(100):
    array = V_3d[:, i, :]
    V_new[i] = array.max(axis=1)
    W_index = np.argmax(array, axis=1)
    W_prime[i] = W[W_index]
    
plt.plot(W,np.average(W_prime ,axis=1))
plt.title("Policy Function")
plt.show()    
```
